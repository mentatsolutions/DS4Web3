#### Imports
Import the necessary libraries. 

```python
from IPython.display import Image
from IPython.core.display import HTML
from matplotlib import pyplot as plt
import requests
import urllib.request
import time
```

#### Prompt
To make a request to the website specified we need a prompt. We will pass the prompt to the Leonardo API in the payload.

```python
prompt='Wild book. An ancient animal book, its pages filled with knowledge and wisdom, its cover alive with intricate designs. The book is closed with a belt'
```

#### API request
Use the requests libraries to make a request to the `generations` API endpoint. In the payload we pass the characteristics of the image to be generated. 
The `last_generation` variable retrieves the work ID. This will be useful to retrieve the images in the subsequent cells.

```python
url = "https://cloud.leonardo.ai/api/rest/v1/generations"

payload = {
    "prompt": prompt,
    "modelId":"6bef9f1b-29cb-40c7-b9df-32b51c1f67d3",
    "width": 512,
    "height": 512,
    "sd_version": "v2",
    "presetStyle": "LEONARDO",
     "public": False,
    "promptMagic": True
}
headers = {
    "accept": "application/json",
    "content-type": "application/json",
    "Authorization": "[YOUR API KEY]"
}

response = requests.post(url, json=payload, headers=headers)
print (response.text)

data=response.json()
last_generation= data['sdGenerationJob']['generationId']
```

Use the `time` library to pause the execution of the notebook for 15 seconds. This is necessary to allow the image to be generated by the website before attempting to retrieve it.

```python
time.sleep(15) 
```

Defining a prefix `wildbook_generated_` for the images to be retrieved. 

```python
result_prefix = "wildbook_generated_"
```

#### Retrieve and show the Images
Use the `requests` library to make a request to the `leonardo.ai` API and retrieve a list of images. Loop through the list of images and download each one to the local directory, using the `urllib` library. The images will be saved with the prefix `wildbook_generated_` and a numerical suffix.

```python
url = "https://cloud.leonardo.ai/api/rest/v1/generations/"+last_generation

headers = {"accept": "application/json",
           "Authorization": "[YOUR API KEY]"}

response = requests.get(url, headers=headers)
images=response.json()
```

```python
for pic in range(len(images['generations_by_pk']['generated_images'])):
    url=images['generations_by_pk']['generated_images'][pic]['url']
    fname = result_prefix + str (pic)+".jpg" 
    urllib.request.urlretrieve(url, fname)
    display (Image(url=url))
```
